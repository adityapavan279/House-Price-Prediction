###  HYPOTHESIS TESTING

```python
housing.head()
```

```python
sfur=housing[housing.furnishingstatus=='semi-furnished']
ufur=housing[housing.furnishingstatus=='unfurnished']
ffur=housing[housing.furnishingstatus=='furnished']
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(sfur.price,ufur.price,ffur.price)
# Since p-value:2.0928668857953132e-12 is less than 0.05 we reject null hypothesis
# means there is no significant difference betwee price and furnishingstatus.
```

### Scaling Data

```python
housing.head()
```

```python
objcols=housing.select_dtypes(include='object')
numcols=housing.select_dtypes(include=np.number)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
objcols_scaled=objcols.apply(le.fit_transform)
```

```python
objcols_scaled
```

### Model Building

```python
y=housing.price
```

```python
numcols=numcols.drop('price',axis=1)
```

```python
combine_housing=pd.concat([numcols,objcols_scaled],axis=1)
```

```python
x=combine_housing
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
regmodel=reg.fit(x,y)
```

```python
regmodel.score(x,y)
```

```python
regpred=regmodel.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(regpred,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree=DecisionTreeRegressor(max_depth=8)
```

```python
treemodel=tree.fit(x,y)
```

```python
treemodel.score(x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rfmodel=rf.fit(x,y)
```

```python
rfmodel.score(x,y)
```
